# Chi - Square Distribution 

```python
from scipy.stats import norm
import pandas
import numpy as np
import matplotlib.pylab as pl
%matplotlib inline

### a Chi-Squared distribution ###

mymean = 100
df = mymean
chi = np.random.chisquare(df, size = 100)
pl.hist(chi, normed=True, bins = 50)
pl.ylabel("N")
pl.xlabel("x")



```

```python
# to get the mean and std if this chi-squared dist
print ("Chisq mean: %.2f, standard deviation: %.2f"%(chi.mean(), chi.std()))
```

```python
sampsize = np.linspace(10, 2000, 100).astype(int)
print (sampsize)
```

```python
chi = {}
for n in sampsize:
    chi[n] = np.random.chisquare(df, size = n)
    
chi["means"] = {}
axchi_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

#print (chi['means'])
### not sure what is that mean?
for nn in chi:
    if not type(nn) == str:
        chi['means'][nn] = chi[nn].mean()

#print (chi['means'][nn], chi[nn].mean())
    #plot:
        axchi_mu_n.plot(nn, chi['means'][nn], 'o')
        axchi_mu_n.set_xlabel('sample size', fontsize=18)
        axchi_mu_n.set_ylabel('sample mean', fontsize=18)
        axchi_mu_n.set_title('Chi squared', fontsize=18)
        axchi_mu_n.plot([min(sampsize), max(sampsize)], [df, df], 'k')

```

```python
#get all the means in the list
allmeans = list(chi["means"].values())
#plot into histogram 
pl.figure(figsize=(6, 6))
pl.hist(allmeans,normed=True,bins= 30, color = "g")
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)

x_min,x_max = pl.xlim()
#fit a normal as gaussian
mu, std = norm.fit(allmeans)
x = np.linspace(x_min, x_max, 100)

p = norm.pdf(x, mu, std)
pl.plot(x, p, 'k', linewidth=2, label="Fitted Gaussian",linestyle= "--")
title = "Fit results: mu = %.2f,  std = %.2f" % (mu, std)

 
```

# Do it for the binominal distribution
